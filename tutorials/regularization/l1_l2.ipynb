### 2d 等高线

```python
Image(url='https://miro.medium.com/v2/resize:fit:640/format:webp/0*IgQLm-qC3bBLMHc5.png', width=400)
```

- L1正则（左图）：等高线是由绝对值函数定义的菱形。
    - L1更倾向于产生稀疏解
    - L1正则的约束边界是菱形，其顶点与坐标轴交点重合。例如，在二维平面中，L1约束的菱形有四个顶点分别位于 $(\pm r,0), (0,\pm r)$
    - 梯度下降优化中，目标函数的最优点更容易落在这些顶点上，而不是边界的其他部分。这是因为 L1 正则的梯度在菱形的边界附近呈现“尖锐变化”，使得优化路径倾向于指向这些稀疏的顶点。
- L2正则（右图）：等高线是由二次函数定义的同心圆。

```python
# Define L1 and L2 regularization functions
def l1_regularization(x, y):
    return np.abs(x) + np.abs(y)

def l2_regularization(x, y):
    return x**2 + y**2
```

```python

# Generate grid points
x = np.linspace(-2, 2, 400)
y = np.linspace(-2, 2, 400)
X, Y = np.meshgrid(x, y)

# Calculate L1 and L2 values
Z_l1 = l1_regularization(X, Y)
Z_l2 = l2_regularization(X, Y)

# Plot L1 regularization
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.contour(X, Y, Z_l1, levels=10, cmap="viridis")
plt.title("L1 Regularization Contours")
plt.xlabel("x")
plt.ylabel("y")
plt.axis("equal")

# Plot L2 regularization
plt.subplot(1, 2, 2)
plt.contour(X, Y, Z_l2, levels=10, cmap="viridis")
plt.title("L2 Regularization Contours")
plt.xlabel("x")
plt.ylabel("y")
plt.axis("equal")

plt.tight_layout()
```
